1. Load the Palmer Penguins dataset:

```python
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
df.head()
df.info()
```

2. The dataset contains missing values. Handle them by removing rows with NaN values:

```python
df_clean=df.dropna()
df_clean.info()
```

# Part 2: Data Visualization

Task 1: Boxplot - Analyzing Body Mass by Species
Create a boxplot to compare the distribution of penguin body mass across different species:

```python
plt.figure(figsize=(10,8))
sns.set_theme(style="darkgrid")
sns.boxplot(x='species', y='body_mass_g', data=df_clean, color="#3D562F")
plt.title('Body Mass by Species', fontsize=16)
plt.xlabel('Penguin Species', fontsize=14)
plt.ylabel('Body Mass (g)', fontsize=14)
plt.tick_params(labelsize=12)

plt.show()
```

Interpretation: Based on the boxplot, answer the following:

1. Which species has the largest median body mass?
2. Which species has the widest range in body mass?
3. Are there any outliers? If so, which species have them?

Answers:

1. The Gentoo Penguin has the largest median body mass.
2. The Gentoo also seem to have the widest range in body mass.
3. Only the mass distribution for the Chinstrap Penguins show any outliers.

Task 2: Violin Plot - Flipper Length Distribution by Species
Create a violin plot to analyze the distribution of flipper length across species:

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='species', y='flipper_length_mm', data=df_clean, color="#3D562F",inner='quartile',linecolor='white')
plt.title('Distribution of Flipper Lengths (mm) per Species',fontsize=16)
plt.xlabel('Species',fontsize=14)
plt.ylabel('Mean Flipper Length (mm)',fontsize=14)
plt.tick_params(labelsize=12)
plt.show()
```

Interpretation: Answer the following:

1. Which species has the most symmetrical distribution?
2. Which species has the widest spread of flipper lengths?
3. How do the quartiles (white lines in the plot) compare for each species?

Answers:
1. The Adelie Penguins have the most symmetrical distribution.
2. The Adelie also seem to have the widest flipper length spread.
3. Adelies have lower median flipper length than the other species, with ~50% of the distribution for Chinstrap penguins falling above the third quartile for the Adelie. The Gentoo have the highest median flipper length, with nearly 100% of their data falling above the third quartile of the Chinstrap distribution. While the Adelie and Chinstrap penguins have apparently normal flipper length distributions, the Gentoo have a distribution that is a little less normal, showing a higher density represented around 230 mm flipper length.
