#Week 3: Matplotlib Line and Scatter Plots Assignment

1. Setup and Styles

```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
```

2. Basic Line Plot

```python
x=np.linspace(-2,4,50)
y = x**2 -2*x + 1
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.show()
```

3. Multiple lines

```python
x=np.linspace(-5,10,50)
y=x**3
g=np.exp(x)
h=np.log(x)
plt.figure(figsize=(10,8))
plt.ylim(-100,1000)
plt.plot(x,y, color='green',label="x^3")
plt.plot(x,g, color='blue',label='e^x')
plt.plot(x,h,color='orange',label='log(x)')
plt.legend()
plt.show()
```

4. Styling Lines

```python
x=np.linspace(-10,10,100)
y=x**2
g=x**3
h=x**4
z=x**5
plt.figure(figsize=(10,8))
plt.ylim(-100,100)
plt.plot(x,y, color='green',label="x^2",linestyle='solid')
plt.plot(x,g, color='b',label='x^3',linestyle=':')
plt.plot(x,h,color='.25',label='x^4',linestyle='--')
plt.plot(x,z,color='#FD3DB5',label='x^5',linestyle='dashdot')
plt.legend()
plt.show()
```

5. Axes Limits and Scaling

```python
x=np.linspace(-10,10,100)
y=1/(x**2+1)
plt.plot(x,y)
plt.ylim(-.5,1.5)
plt.xlim(-5,5)
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=1/(x**2+1)
plt.plot(y,x)
plt.axis('tight')
plt.show()
```

6. Labels and Legends

```python
x=np.linspace(0,10,50)
y=np.sqrt(x)
g=x**(1/3)
plt.figure(figsize=(10,8))
plt.plot(x,y, color='indigo',label="sqrt(x)")
plt.plot(x,g, color='red',label='x^(1/3)')
plt.title("Plot of Square Root X and X^1/3")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

7. Basic Scatter Plot

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.figure(figsize=(10,8))
plt.plot(x,y,marker='o')
plt.title('Plot using plt.plot')
plt.show()
plt.figure(figsize=(10,8))
plt.scatter(x,y,marker='o')
plt.title('Plot using plt.scatter')
plt.show()
```

8. Customizing Scatter Plots

```python
x=np.linspace(0,np.pi*4,50)
y=x*np.sin(x)
plt.figure(figsize=(10,8))
plt.plot(x,y,marker='o',
         markersize=7,
         markerfacecolor='lightblue',
         linewidth=2, color='indigo',
         markeredgecolor='darkblue',
         markeredgewidth=.5)
plt.show()
```

9. Combined Plot

```python
x_line=np.linspace(-2*np.pi,2*np.pi,50)
y_line=x_line*np.cos(x_line)
x_scatter=np.arange(-2*np.pi,2*np.pi,np.pi/4)
y_scatter=x_scatter*np.cos(x_scatter)
plt.figure(figsize=(10,8))
plt.plot(x_line,y_line, color='grey',linestyle='dashed')
plt.scatter(x_scatter,y_scatter, color='green',alpha=.70,marker='s')
plt.show()
```

10. Custom Plot with Alternative Style

```python
plt.style.use('ggplot')
x_line=np.linspace(-2*np.pi,2*np.pi,100)
x_scatter=np.arange(-2*np.pi,2*np.pi,np.pi/8)
y_line1=np.tan(x_line)
y_scatter1=np.tan(x_scatter)
y_line2=np.arctan(x_line)
y_scatter2=np.arctan(x_scatter)
plt.figure(figsize=(10,8))
plt.ylim(-10,10)
plt.plot(x_line,y_line1,
         color='blue',
         label='tan(x)',
         linewidth=2,
         linestyle='solid')
plt.scatter(x_scatter,y_scatter1,
            marker='^',
            s=60,
            color='lightblue',
            edgecolor='blue')
plt.plot(x_line,y_line2,
         color='darkorange',
         label='arctan(x)',
         linewidth=2,
         linestyle='dotted')
plt.scatter(x_scatter,y_scatter2,
            s=60,
            color='#FFCE1B',
            marker='o',
            edgecolor='darkorange')
plt.legend()
plt.show()
```

###Comments:
I chose to graph tan(x) and arctan(x), because they have an actual relationship with arctan essentially being the inverse of tan. I thought that would make it easier to show them both on the same range on a graph.

Because they are inverse to each other, I wanted to pick completely opposite colors for both of them so I chose orange for arctan and blue for tan, with the markers essentially being a lighter version of each color.

ggplot's main difference to seaborn whitegrid is that it has a grey background, which I think can be helpful for seeing lighter colors. The axis values are also grey by default, but I think I'd manually change them to black to make them stand out more.

11. Style Comparison

```python
plt.style.use('default')
x=np.linspace(0,np.pi*4,50)
y=x*np.sin(x)
plt.figure(figsize=(10,8))
plt.plot(x,y,marker='o',
         markersize=7,
         markerfacecolor='lightblue',
         linewidth=2, color='indigo',
         markeredgecolor='darkblue',
         markeredgewidth=.5)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
x=np.linspace(0,np.pi*4,50)
y=x*np.sin(x)
plt.figure(figsize=(10,8))
plt.plot(x,y,marker='o',
         markersize=7,
         markerfacecolor='lightblue',
         linewidth=2, color='indigo',
         markeredgecolor='darkblue',
         markeredgewidth=.5)
plt.show()
```

```python
plt.style.use('ggplot')
x=np.linspace(0,np.pi*4,50)
y=x*np.sin(x)
plt.figure(figsize=(10,8))
plt.plot(x,y,marker='o',
         markersize=7,
         markerfacecolor='lightblue',
         linewidth=2, color='indigo',
         markeredgecolor='darkblue',
         markeredgewidth=.5)
plt.show()
```

###Comments:
All three styles look pretty nice. The default style does not have a grid, but the default font size for the axes values is comparable to seaborn-whitegrid and ggplot. They could all reasonably be used for this graph, but I think for this particular one the seaborn-whitegrid would be the best style because it includes a grid that makes it easier to track the graphed line but it doesn't have ggplot's grey background, which allows the graphed function to actually pop-out more. The axes values are also darker than ggplot's which I also think makes them easier to read.
