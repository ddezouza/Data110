# Week 4: Crime Data Visualization (Bar charts)

#### Setup:

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

#### Dataset:

```python
Offenses = ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape','Robbery']
Of_2023 = [264.1,250.7,1347.2,318.7,5.7,38.0,66.5] #Offenses per 100,000 people (2023)
Delta = [-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8] #Change in offenses per 100,000 people (2022-2023)
Of_2022 = [267.4,258.8,1352.1,306.7,17.7,47.8,67.3] #For combined bar chart
```

```python
data = {'Offenses': ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape','Robbery'],
        'Of_2023': [264.1,250.7,1347.2,318.7,5.7,38.0,66.5],
        'Delta':[-3.3,-8.1,-4.9,12.0,-12.0,-9.8,-0.8],
        'Of_2022':[267.4,258.8,1352.1,306.7,17.7,47.8,67.3] }
df = pd.DataFrame(data)
print(df)
```

1. Basic Bar Chart for Crime Rates in 2023

```python
plt.figure(figsize=(10, 8))
plt.bar(df['Offenses'], df['Of_2023'], color='plum',zorder=2)
plt.title('Criminal Offenses per 100,000 People (2023)', fontsize=14)
plt.xlabel('Offenses',fontsize=12)
plt.ylabel('Number of Offenses per 100,000 People',fontsize=12)
plt.grid(axis='y',linestyle='--',zorder=0)
ax = plt.gca()
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()
```

2. Change in Crime Rates (2022 vs. 2023) with Unique Colors

```python
plt.figure(figsize=(10, 8))
colors = ['violet', 'plum', 'orchid', 'indigo', 'lavender','pink','purple']
plt.barh(df['Offenses'], df['Delta'], color=colors,zorder=2)
plt.title('Change in Criminal Offenses per 100,000 People (2022-2023)', fontsize=14)
plt.xlabel('Change in Offenses per 100,000 People',fontsize=12)
plt.ylabel('Offenses',fontsize=12)
plt.grid(axis='x',linestyle='--',zorder=0)
ax = plt.gca()
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()
```

#### Comments:
Absolute numbers tend to be pretty useless unless you already have a background on the topic that is being looked into. I could say there are 1000 axolotls in a particular lake in Mexico but without appropriate context and an intimation as to why that number is important the information is meaningless.

For statistics for things such as crime, absolute numbers without appropriate context can be used for alarmist publications and fail to actually inform the public.

The change statistic has some built in context - it is already a comparison of the same element in a previous time period, and conveys more information than the absolute number.

```python
#Alternate 1: Change orientation + color to better communicate increase/decrease
plt.figure(figsize=(10, 8))
colors = ['green', 'lightgreen', 'chartreuse', 'red', 'lime','yellowgreen','darkgreen']
plt.bar(df['Offenses'], df['Delta'], color=colors,zorder=2)
plt.title('Change in Criminal Offenses per 100,000 People (2022-2023)', fontsize=14)
plt.xlabel('Offenses',fontsize=12)
plt.ylabel('Change in Offenses per 100,000 People',fontsize=12)
plt.grid(axis='x',linestyle='--',zorder=0)
ax = plt.gca()
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()
```

```python
#Alternate 2: Ordering by Delta; altering colors to communicate Increase/Decrease
df_sorted=df.sort_values(by='Delta', ascending=True)
plt.figure(figsize=(12, 10))
colors = ['#005a32', '#238b45', '#41ab5d', '#74c476', '#a1d99b','#c7e9c0','#ba3c3c']
plt.bar(df_sorted['Offenses'], df_sorted['Delta'], color=colors,zorder=2)
plt.title('Change in Criminal Offenses per 100,000 People (2022-2023)', fontsize=14)
plt.xlabel('Offenses',fontsize=12)
plt.ylabel('Change in Offenses per 100,000 People',fontsize=12)
plt.grid(axis='y',linestyle='--',zorder=0)
ax = plt.gca()
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()
```

#### Optional: Combined Bar Chart for Crime Rate in 2022 and 2023

```python
#2022 crime rates already calculated in dataset block
plt.figure(figsize=(12, 10))
x_indexes = np.arange(len(df['Offenses']))
plt.bar(x_indexes - 0.125, df['Of_2022'], width=0.25, label='2022',color='#DDA0DD',zorder=2)
plt.bar(x_indexes + 0.125, df['Of_2023'], width=0.25, label='2023',color='#662d91',zorder=2)
plt.xticks(ticks=x_indexes, labels=df['Offenses'])
plt.legend()
plt.title('Criminal Offenses per 100,000 People (2022-2023)', fontsize=14)
plt.xlabel('Offenses',fontsize=12)
plt.ylabel('Offenses per 100,000 People',fontsize=12)
plt.grid(axis='y',linestyle='--',zorder=0)
ax = plt.gca()
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
plt.show()
```
