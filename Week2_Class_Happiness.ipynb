```python
#maatplot functions
#Happier countries have lower ranks
#First create canvas then plot
plt.figure(figsize=[16,10])
plt.scatter(df['Rank'],df['HappinessScore'],s=80,color='red',marker='v')
plt.xlabel('Rank')
plt.ylabel('Happiness Score')
plt.title("Rank vs Happiness")
#plt.grid()
plt.show()
```

```python
plt.figure(figsize=[16,10])
plt.scatter(df['HappinessScore'],df['Social support'],s=40,c='purple',marker='s')
plt.xlabel('Happiness Score')
plt.ylabel('Social Support')
plt.title("Social Support vs Happiness")
plt.show()
```

```python
plt.figure(figsize=[16,10])
plt.scatter(df['Generosity'],df['Social support'],s=120,marker='*')
plt.xlabel('Generosity')
plt.ylabel('Social Support')
plt.title('Generosity vs. Social Support')
```
